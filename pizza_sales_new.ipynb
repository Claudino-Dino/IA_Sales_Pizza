```python
# IMPORTAR BASE DE DADOS
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from collections import Counter

tabela_pizza = pd.read_csv('pizza_sales_treino_dow.csv')
```

```python
pizza = 'The Classic Deluxe Pizza'
tabela_pizza[pizza] = tabela_pizza.apply(lambda row: row["quantity"] if row["pizza_name"] == pizza else 0, axis=1)
```

```python
# REMOVER COLUNAS DESNECESSÁRIAS:
colunas_rmv = [
    'total_price',
    'unit_price',
    'pizza_size',
    'pizza_category',
    'pizza_ingredients'
]
```

```python
tabela_pizza_reduzido = tabela_pizza.drop(columns=colunas_rmv)
```

```python
# CODIFICAR COLUNAS QUE SÃO "Object":

codificador = LabelEncoder()
tabela_pizza_reduzido["pizza_name_id"] = codificador.fit_transform(tabela_pizza_reduzido["pizza_name_id"])
tabela_pizza_reduzido["order_date"] = codificador.fit_transform(tabela_pizza_reduzido["order_date"])
tabela_pizza_reduzido["order_time"] = codificador.fit_transform(tabela_pizza_reduzido["order_time"])
tabela_pizza_reduzido["pizza_name"] = codificador.fit_transform(tabela_pizza_reduzido["pizza_name"])
tabela_pizza_reduzido["days_of_week"] = codificador.fit_transform(tabela_pizza_reduzido["days_of_week"])
```

```python
# SEPARAR TREINO DE TESTE:

y = tabela_pizza_reduzido["days_of_week"]

x = tabela_pizza_reduzido

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
# UTILIZAR OS MODELOS DE IA:
modelo_decision_tree = RandomForestClassifier()
```

```python
# TREINAR IA's:
modelo_decision_tree.fit(x_treino, y_treino)
```

```python
# VERIFICAR PRECISÃO DAS IA's:
previsao_arvore_decisao = modelo_decision_tree.predict(x_teste)

print(accuracy_score(y_teste, previsao_arvore_decisao))
```

```python
# PREVER CASO REAL; SE A PIZZA ESPECÍFICA TIVER MAIS PEDIDOS NESSE DIA:

contagem = 0

contagem = Counter(previsao_arvore_decisao)


converter_num_dia = {

    0: "Domingo",
    1: "Segunda-Feira",
    2: "Terça-Feira",
    3: "Quarta-Feira",
    4: "Quinta-feira",
    5: "Sexta-Feira",
    6: "Sabado"
}

top_1_dia, frequencia = contagem.most_common(1)[0]

print(f"O dia que mais vende a pizza '{pizza},' é: ** {converter_num_dia[top_1_dia]} **\nE a quantidade de vendas dessa pizza nesse dia ao longo do ano de 2015 é: {frequencia}")
```
